# PHẦN 1 - DATA CLEANING
Mục tiêu:
- Đọc dữ liệu từ file CSV
- Kiểm tra missing values
- Kiểm tra duplicate rows
- Xử lý giá trị thiếu
- Chuyển OrderDate sang datetime
- Tạo cột Month từ OrderDate

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Read CSV File

```python
# Đọc dữ liệu từ file CSV
# Đổi tên file nếu bạn lưu với tên khác
df = pd.read_csv(r"E:\AIO\Project\sales_forecasting_xai\shared\notebooks\middle_exam\sales_data5.csv")

# Xem 5 dòng đầu
df.head()
```

## Overview Dataset

```python
# Kiểm tra kích thước dữ liệu và kiểu dữ liệu
print("Shape:", df.shape)
df.info()
```

## Check Missing Values

```python
# Kiểm tra số lượng giá trị thiếu theo từng cột
missing_values = df.isnull().sum()

print("Missing values by column:")
print(missing_values)
```

## Check Duplicate Rows

```python
# Kiểm tra số dòng bị trùng
duplicate_count = df.duplicated().sum()

print("Number of duplicate rows:", duplicate_count)

# Hiển thị các dòng trùng nếu có
df[df.duplicated()]
```

## Handle Missing Values

```python
df = df.dropna().copy()

print("Shape after handling missing values:", df.shape)
print(df.isnull().sum())
```

## Convert OrderDate to Datetime

```python
# Chuyển cột OrderDate sang datetime
df["OrderDate"] = pd.to_datetime(df["OrderDate"], errors="coerce")

# Kiểm tra lại kiểu dữ liệu
print(df.dtypes)
```

## Recheck Missing After Datetime Conversion

```python
# Sau khi chuyển datetime, kiểm tra lại vì dữ liệu lỗi có thể thành NaT
print("Missing values after datetime conversion:")
print(df.isnull().sum())
```

## Remove Invalid Dates If Any

```python
# Xóa các dòng có OrderDate bị lỗi sau khi convert
df = df.dropna(subset=["OrderDate"]).copy()

print("Shape after removing invalid OrderDate:", df.shape)
```

## Create Month Column

```python
# Tạo cột Month từ OrderDate
df["Month"] = df["OrderDate"].dt.month

df.head()
```

## Final Cleaned Data Check

```python
# Kiểm tra dữ liệu sau khi làm sạch
print("Final shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

df.head()
```

## Save Cleaned Data

```python
# Lưu dữ liệu đã làm sạch ra file mới
df.to_csv("orders_cleaned.csv", index=False)
```
