# PHẦN 5 - ADVANCED
## Câu 1: Tạo dashboard

Dashboard đã được xây dựng để trực quan hóa các chỉ số kinh doanh chính, bao gồm:
- Tổng doanh thu
- Tổng lợi nhuận
- Tổng số lượng bán
- Doanh thu theo tháng
- Doanh thu theo khu vực
- Tỷ lệ doanh thu theo Category
- Mối quan hệ giữa Sales và Profit
- Phân tích theo CustomerType và xu hướng theo thời gian

Dashboard giúp người dùng theo dõi dữ liệu tổng quan trên cùng một giao diện, từ đó hỗ trợ phân tích và ra quyết định nhanh hơn.

## Câu 2: Phân tích CustomerType và xu hướng theo thời gian

### 1. Doanh thu theo CustomerType
Dựa trên dashboard, nhóm khách hàng Consumer mang lại doanh thu cao hơn Corporate. Cụ thể, Consumer đạt khoảng 71k USD, trong khi Corporate đạt khoảng 62k USD.

Điều này cho thấy nhóm Consumer hiện là tệp khách hàng đóng góp doanh thu chính cho doanh nghiệp. Vì vậy, doanh nghiệp nên ưu tiên duy trì và mở rộng nhóm khách hàng này.

### 2. Xu hướng doanh thu theo thời gian
Biểu đồ xu hướng theo tháng cho thấy doanh thu của cả Consumer và Corporate đều biến động theo thời gian, không tăng đều liên tục.

Nhóm Consumer giữ mức doanh thu cao hơn trong nhiều tháng và có nhiều giai đoạn tăng mạnh, đặc biệt vào khoảng tháng 4, tháng 6, tháng 9 và tháng 10. Trong khi đó, nhóm Corporate cũng tăng ở một số thời điểm nhưng biến động mạnh hơn và giảm đáng kể vào cuối năm, nhất là tháng 12.

### 3. Nhận xét chung
- Consumer là nhóm khách hàng có giá trị doanh thu cao hơn.
- Cả hai nhóm đều có xu hướng biến động theo tháng.
- Consumer ổn định hơn và có tiềm năng tiếp tục khai thác.
- Corporate cần được cải thiện ở các giai đoạn doanh thu thấp, đặc biệt cuối năm.
