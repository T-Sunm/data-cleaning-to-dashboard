# PHẦN 4 - INSIGHT & BUSINESS ANALYSIS

## 1. Khu vực nào mang lại doanh thu cao nhất?
Dựa trên biểu đồ doanh thu theo khu vực, South là khu vực có doanh thu cao nhất, đạt khoảng 36,324.08. Trong khi đó, North là khu vực có doanh thu thấp nhất.

## 2. Category nào doanh thu thấp nhất?
Từ biểu đồ tỷ lệ doanh thu theo Category, Office Supplies là nhóm có doanh thu thấp nhất, chiếm khoảng 32.1% tổng doanh thu. Hai nhóm còn lại là Technology và Furniture có tỷ trọng cao hơn.

## 3. Có mối quan hệ giữa Sales và Profit không?
Biểu đồ scatter plot cho thấy Sales và Profit có xu hướng đồng biến, tức là khi doanh thu tăng thì lợi nhuận thường cũng tăng theo. Tuy nhiên, mối quan hệ này không hoàn toàn chặt chẽ vì có nhiều điểm dữ liệu phân tán và vẫn xuất hiện các đơn hàng doanh thu cao nhưng lợi nhuận thấp hoặc âm.

## 4. Đề xuất 2 chiến lược cải thiện doanh thu
- Tăng đầu tư vào khu vực South và áp dụng chiến lược bán hàng hiệu quả của South sang các khu vực doanh thu thấp hơn như North.
- Đẩy mạnh category Office Supplies bằng các chương trình combo, upsell hoặc cross-sell với những nhóm sản phẩm có doanh thu tốt hơn để tăng giá trị đơn hàng.
