# PHẦN 2 - DATA EXPLORATION
Mục tiêu:
- Tính tổng doanh thu, tổng lợi nhuận, tổng số lượng bán
- Tìm top 5 sản phẩm có doanh thu cao nhất
- Phân tích doanh thu theo Region và Category

## Load Cleaned Data

```python
import pandas as pd

# Đọc dữ liệu đã làm sạch từ phần 1
df = pd.read_csv("orders_cleaned.csv")

# Nếu cần, chuyển lại OrderDate sang datetime
df["OrderDate"] = pd.to_datetime(df["OrderDate"], errors="coerce")

df.head()
```

## Câu 1: Tổng doanh thu, lợi nhuận, số lượng

```python
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_quantity = df["Quantity"].sum()

print("Tổng doanh thu:", total_sales)
print("Tổng lợi nhuận:", total_profit)
print("Tổng số lượng bán:", total_quantity)
```

## Câu 1: Summary Table

```python
summary_metrics = pd.DataFrame({
    "Metric": ["Total Sales", "Total Profit", "Total Quantity"],
    "Value": [df["Sales"].sum(), df["Profit"].sum(), df["Quantity"].sum()]
})

summary_metrics
```

## Câu 2: Top 5 sản phẩm doanh thu cao nhất

```python
top_5_products = (
    df.groupby("ProductName", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
      .head(5)
)

top_5_products
```

## Câu 2: Top 5 sản phẩm kèm tỷ trọng doanh thu

```python
top_5_products = top_5_products.copy()
top_5_products["Sales_Percentage"] = (
    top_5_products["Sales"] / df["Sales"].sum() * 100
).round(2)

top_5_products
```

## Câu 3: Doanh thu theo Region

```python
sales_by_region = (
    df.groupby("Region", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

sales_by_region
```

##  Câu 3: Doanh thu theo Category

```python
sales_by_category = (
    df.groupby("Category", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

sales_by_category
```

## Summary

```python
sales_by_category = (
    df.groupby("Category", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

sales_by_category
```

```python
print("=== Tổng quan chỉ số ===")
print(summary_metrics)

print("\n=== Top 5 sản phẩm doanh thu cao nhất ===")
print(top_5_products)

print("\n=== Doanh thu theo Region ===")
print(sales_by_region)

print("\n=== Doanh thu theo Category ===")
print(sales_by_category)
```
